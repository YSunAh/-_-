```python
import csv
f = open('gender_01.csv')
data = csv.reader(f)
next(data)
m = []
f= []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i.replace(',','')))
        for i in row[106:] :
            f.append(int(i.replace(',','')))
        break

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.scatter(m,f,c=range(101) ,alpha = 0.5, cmap = 'jet')
plt.colorbar()
plt.plot(range(max(m)),range(max(m)), 'g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
    
plt.title('p150_Bubble Chart')
plt.scatter(x,y,s=size, c=range(100) , cmap = 'jet',alpha= 0.7 )
plt.colorbar()
plt.show()
```

```python
import csv
f = open('gender_01.csv')
data = csv.reader(f)

m = []
f= []


for row in data :
    if '제주특별자치도'in row[0]:
        for i in range(3,104) :
            m.append(int( row[i].replace(',','')))
            f.append(int( row[i+103].replace(',','')))
        break
        
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5),dpi = 300)
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('Oct_2019_Population_제주특별자치도 남성-여성 scatter Chart')
plt.xlabel('남성')
plt.ylabel('여성')

plt.scatter(m,f,c=range(101), cmap='jet',alpha=0.7)
plt.plot(range(max(m)),range(max(m)), 'g')
plt.colorbar()
plt.show()
```
