```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

high=[]
low=[]


    
for row in data :
    if row[-1] != ''and row[-2] !='':
        if 1920 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '01'and row[0].split('-')[2]=='31':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
            
print('max Temp :',high)
print()
print('max Temp :',low)

plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False

plt.title('My Birthday(31/01/2021)LowHigh Temp in Seoul')
plt.xlabel('date')
plt.ylabel('Temp(Celsious)')

plt.style.use('ggplot')

red_square = dict(markerfacecolor = 'r',marker='s')
blue_diamond = dict(markerfacecolor = 'b',marker = 'D')


plt.boxplot([low, high], flierprops=red_square)

plt.xticks([1,2],labels = ['최저온도','최고온도'])
plt.show()
```
