```python
import csv
f = open('subwaytime_01.csv')
data = csv.reader(f)
next(data)
next(data)

onMaxNoOfPassenger = [0]*24
onMaxStation = ['']*24
offMaxNoOfPassenger = [0]*24
offMaxStation = ['']*24

for row in data:
    row[4:] = map(int,row[4:])
    
    for j in range(24):
    
        tempPassenger = row[j*2+4]
        if tempPassenger > onMaxNoOfPassenger[j]:
            onMaxNoOfPassenger[j] = tempPassenger
            onMaxStation[j] = row[3] + '(' + str(j+4) + ')'
    
    
        tempPassenger = row[j* 2+4+1]
        if tempPassenger > offMaxNoOfPassenger[j]:
            offMaxNoOfPassenger[j] = tempPassenger
            offMaxStation[j] = row[3] + '(' + str(j+4) + ')'
            
print('승차: ', onMaxNoOfPassenger)
print('하차: ', offMaxNoOfPassenger)
            
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize = (10,3))
plt.rc('font',family = 'Malgun Gothic')
plt.title('시간대별로 최대 승차인원')

plt.bar(range(24), onMaxNoOfPassenger, color = 'hotpink', label='승차')
plt.xticks(range(24), onMaxStation, rotation = 45)
plt.savefig('images/'+'시간대별로 최대 승차인원'+'.png',bbox_inches='tight')

plt.figure(figsize = (10,3))
#plt.rc('font',family = 'Malgun Gothic')
plt.title('시간대별로 최대 하차인원')
        
plt.bar(range(24), offMaxNoOfPassenger, color = 'skyblue', label='하차')
plt.xticks(range(24), offMaxStation, rotation = 45)
plt.savefig('images/' + '시간대별로 최대 하차인원' + '.png' )#, bbox_inches='tight')
plt.show()
```
