```python
import csv
f = open('age.csv')
data = csv.reader(f)
result = []

name = input('인구구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력하세요: ')
for row in data :
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))
            
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(10,3))
plt.rc('font',family='Malgun Gothic')

plt.title("Oct_2019_Population_"+name)
plt.xlabel('나이')
plt.ylabel('명')
plt.plot(result)
plt.show()
```
